```python
import json
from collections import Counter
import matplotlib.pyplot as plt
from math import nan
```

```python
with open("updated_measurement_conditions.json", "r") as infile:
    updated_measurement_conditions = json.load(infile)
```

```python
ph = []

for key in updated_measurement_conditions.keys():
    loc = updated_measurement_conditions[key]
    ph.append(loc["ph"]) 
```

```python
ph
```

```python
Counter(ph).most_common()
```

```python
ph_as_floats = []
for t in ph:
    if t == "NULL; NULL; NULL":
        ph_as_floats.append(nan)
    elif t == "NULL":
        ph_as_floats.append(nan)
    elif t == "NULL; NULL":
        ph_as_floats.append(nan)
    elif t == "6.0; 6.0; 6.4; 6.4; NULL":
        ph_as_floats.append(nan)
    elif t == "0; 0":
        ph_as_floats.append(nan)
    elif t == "5.5; 5.1":
        ph_as_floats.append(5.1)
    elif t == "6.5; 6.5; 6.5; 6.5":
        ph_as_floats.append(6.5)
    elif t == "5.2; 5.1":
        ph_as_floats.append(nan)
    elif t == "4.8; 4.8; 4.67":
        ph_as_floats.append(nan)
    elif t == "7.2; 7.3":
        ph_as_floats.append(7.2)
    elif t == "6.5; 6.5":
        ph_as_floats.append(6.5)
    elif t == "6.0; 6.0":
        ph_as_floats.append(6.0)
    elif t == "5.5; 5.5":
        ph_as_floats.append(5.5)
    else:
        ph_as_floats.append(float(t))
```

```python
plt.figure(figsize=(15, 10))
plt.hist(ph_as_floats, 200, density=False, alpha=0.75, histtype="stepfilled")
plt.show()
```

## Filtering

#### In the following section the bmrd ids with a corresponding PH value of one will be filtered and saved in a list.

```python
keys = list(updated_measurement_conditions)
```

```python
filtered_by_ph = [keys[index] for index, p in enumerate(ph_as_floats) if p >= 5 and p <= 8]
```
