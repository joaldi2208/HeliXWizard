```python
import json
import matplotlib.pyplot as plt
import itertools
```

```python
with open("secondary_structure.json", "r") as infile:
        secondary_structure = json.load(infile)
```

# Q3 state

```python
filtered_ids = []
number_counts_3_state = []
counter = 0
p_count = 0
for bmrb_id, info in secondary_structure.items():
    counts_in_3_state = sum(info["counts_in_3_state"])
    number_counts_3_state.append(counts_in_3_state)
    if counts_in_3_state < 1:
        counter += 1
        print(counter, bmrb_id, counts_in_3_state)
    elif "P" in info["sec_structure_8_state"]:
        #num_p = 0
        #for pp in info["sec_structure_8_state"]:
        #    if pp == "PP":
        #        num_p += 1
        p_count += 1
        #if num_p > 1:
            #print("bmrb: ", bmrb_id, num_p)
        print("brmb: ", bmrb_id)
    else:
        filtered_ids.append(bmrb_id)
print(p_count)
```

## Prolin II helix is in 3 state and 8 state

```python
pairs = []
p_count = 0
h_count = 0
c_count = 0
g_count = 0
t_count = 0
e_count = 0
s_count = 0
b_count = 0
i_count = 0

p_count_ = 0
h_count_ = 0
c_count_ = 0
g_count_ = 0
t_count_ = 0
e_count_ = 0
s_count_ = 0
b_count_ = 0
i_count_ = 0

for bmrb_id, info in secondary_structure.items():
    state_3_symbol = info["sec_structure_3_state"]
    state_8_symbol = info["sec_structure_8_state"]
    if "P" in info["sec_structure_8_state"]:
        p_count += 1
    if "H" in info["sec_structure_8_state"]:
        h_count += 1
    if "-" in info["sec_structure_8_state"]:
        c_count += 1
    if "G" in info["sec_structure_8_state"]:
        g_count += 1
    if "T" in info["sec_structure_8_state"]:
        t_count += 1
    if "E" in info["sec_structure_8_state"]:
        e_count += 1
    if "S" in info["sec_structure_8_state"]:
        s_count += 1
    if "B" in info["sec_structure_8_state"]:
        b_count += 1
    if "I" in info["sec_structure_8_state"]:
        i_count += 1
    for sym3, sym8 in zip(state_3_symbol, state_8_symbol):
        if (sym8, sym3) not in pairs:
            pairs.append((sym8, sym3))
        if sym8 == "P":
            p_count_ += 1 
        if sym8 == "H":
            h_count_ += 1
        if sym8 == "-":
            c_count_ += 1 
        if sym8 == "G":
            g_count_ += 1
        if sym8 == "T":
            t_count_ += 1 
        if sym8 == "E":
            e_count_ += 1
        if sym8 == "S":
            s_count_ += 1 
        if sym8 == "B":
            b_count_ += 1
        if sym8 == "I":
            i_count_ += 1 
            
            
print("Appearence in molcules")
print("Number of P's: ", p_count)
print("Number of H's: ", h_count)
print("Number of G's: ", g_count)
print("Number of T's: ", t_count)
print("Number of E's: ", e_count)
print("Number of S's: ", s_count)
print("Number of B's: ", b_count)
print("Number of C's: ", c_count)
print("Number of I's: ", i_count)
print()
print("overall appearance")
print("Number of P's: ", p_count_)
print("Number of H's: ", h_count_)
print("Number of G's: ", g_count_)
print("Number of T's: ", t_count_)
print("Number of E's: ", e_count_)
print("Number of S's: ", s_count_)
print("Number of B's: ", b_count_)
print("Number of C's: ", c_count_)
print("Number of I's: ", i_count_)
print(pairs)

print()
print(len(list(secondary_structure)))
```
