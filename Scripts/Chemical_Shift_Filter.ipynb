```python
import pickle
import matplotlib.pyplot as plt
```

```python
with open("chemical_shifts.pkl", "rb") as infile:
        chemical_shifts = pickle.load(infile)
```

```python
chemical_shifts[3322]
```

## H-Shifts

```python
plt.figure(figsize=(15,10))
for sheet in chemical_shifts.values():
    plt.hist(sheet.X_shift, 150, density=False, alpha=0.75, histtype="stepfilled", color="black")
plt.title("H-Shifts")
plt.show()
```

## Filtering

```python
H_to_large_or_small = []
for bmrb_id, sheet in chemical_shifts.items():
    if max(sheet.X_shift) > 14 or min(sheet.X_shift) < 4:
        print(bmrb_id, ":", "max --> ",max(sheet.X_shift), "min --> ", min(sheet.X_shift))
        H_to_large_or_small.append(bmrb_id)
```

```python
keys = list(chemical_shifts)
```

```python
len(keys)
```

```python
for bmrb_id in H_to_large_or_small:
    keys.remove(bmrb_id)
```

```python
len(keys)
```

# N-Shifts

```python
plt.figure(figsize=(15,10))
for sheet in chemical_shifts.values():
    plt.hist(sheet.Y_shift, 100, density=False, alpha=0.75, histtype="stepfilled", color="black")
plt.title("N-Shifts")
plt.show()
```

## Filtering

```python
N_to_large_or_small = []
for bmrb_id, sheet in chemical_shifts.items():
    if max(sheet.Y_shift) > 140 or min(sheet.Y_shift) < 90:
        print(bmrb_id, ":", "max --> ",max(sheet.Y_shift), "min --> ", min(sheet.Y_shift))
        N_to_large_or_small.append(bmrb_id)
```

```python
for bmrb_id in N_to_large_or_small:
    if bmrb_id not in H_to_large_or_small:
        print(bmrb_id)
        keys.remove(bmrb_id)
```
