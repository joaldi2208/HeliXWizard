```python
import pickle
import matplotlib.pyplot as plt
```

```python
with open("chemical_shifts.pkl", "rb") as infile:
        chemical_shifts = pickle.load(infile)
```

```python
chemical_shifts
```

```python
chemical_shifts
```

```python
plt.figure(figsize=(15,10))
for sheet in chemical_shifts.values():
    plt.hist(sheet.X_shift, 150, density=False, alpha=0.75, histtype="stepfilled", color="black")
plt.title("H-Shifts")
plt.show()
```

```python
for bmrb_id, sheet in chemical_shifts.items():
    if max(sheet.X_shift) > 100:
        print(bmrb_id, ":", max(sheet.X_shift))
```

```python
plt.figure(figsize=(15,10))
for sheet in chemical_shifts.values():
    plt.hist(sheet.Y_shift, 100, density=False, alpha=0.75, histtype="stepfilled", color="black")
plt.title("N-Shifts")
plt.show()
```

```python
for bmrb_id, sheet in chemical_shifts.items():
    if max(sheet.Y_shift) > 200:
        print(bmrb_id, ":", max(sheet.Y_shift))
```
