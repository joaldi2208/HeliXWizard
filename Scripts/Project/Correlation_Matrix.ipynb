```python
def convert_to_matrix(index_value_pair):
    """
    converts a index value pair to a 10x10 matrix:
    [(x1,y1),(x2,y2),...] --> [[0,0,0,y2,0,...][y1,0,...],...]
    """
    start_vector = np.zeros(100)
    for index, value in index_value_pair:
        start_vector[index] = value
    correlation_matrix = start_vector.reshape(10,10)
    
    return correlation_matrix
    
```

```python
def make_correlation_plot(correlation_matrix, name):
    
    fig, ax = plt.subplots(figsize=(10,6))
    cmap = mpl.colors.ListedColormap(["tab:red","w","tab:blue"])

    bounds = [-1., -0.29, 0.29, 1.]
    norm = mpl.colors.BoundaryNorm(bounds, cmap.N)

    grid_map = ax.imshow(correlation_matrix, interpolation='none', cmap=cmap, norm=norm)
    ax.set_xlim(9.5,-0.5)
    ax.set_ylim(9.5,-0.5)
    
    ax.set_xticks([0.5,1.5,2.5,3.5,4.5,5.5,6.5,7.5,8.5,9.5],[6.5,7,7.5,8,8.5,9,9.5,10,10.5,11])
    ax.set_yticks([0.5,1.5,2.5,3.5,4.5,5.5,6.5,7.5,8.5,9.5],[95,100,105,110,115,120,125,130,135,140])
    
    ax.set_xlabel(r"$^1$H chemical shift (ppm)")
    ax.set_ylabel(r"$^{15}$N chemical shift (ppm)")

    if "coil" in name:
        plt.colorbar(grid_map, label='pearson correlation') # only for coil
    
    plt.grid()
    
    plt.savefig("../../Drafts/Project-Thesis/img/" + name, bbox_inches="tight")
    plt.show()
```

```python
def make_heatmap_plot(correlation_matrix, name):
    
    fig, ax = plt.subplots(figsize=(10,6))
    im = ax.imshow(correlation_matrix, cmap='seismic_r', interpolation="none", vmin=-0.7, vmax=0.7)
    
    ax.set_xlim(9.5,-0.5)
    ax.set_ylim(9.5,-0.5)
    
    ax.set_xticks([0.5,1.5,2.5,3.5,4.5,5.5,6.5,7.5,8.5,9.5],[6.5,7,7.5,8,8.5,9,9.5,10,10.5,11])
    ax.set_yticks([0.5,1.5,2.5,3.5,4.5,5.5,6.5,7.5,8.5,9.5],[95,100,105,110,115,120,125,130,135,140])
    
    ax.set_xlabel(r"$^1$H chemical shift (ppm)")
    ax.set_ylabel(r"$^{15}$N chemical shift (ppm)")
    
    if "coil" in name:
        plt.colorbar(im, label='pearson correlation') # only for coil
    
    plt.grid()
    
    plt.savefig("../../Drafts/Project-Thesis/img/" + name, bbox_inches="tight")
    plt.show()
```
