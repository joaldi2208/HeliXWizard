```python
import json
from collections import Counter
import matplotlib.pyplot as plt
from math import nan
```

```python
with open("updated_measurement_conditions.json", "r") as infile:
    updated_measurement_conditions = json.load(infile)
```

```python
pressure = []

for key in updated_measurement_conditions.keys():
    loc = updated_measurement_conditions[key]
    try:
        pressure.append(loc["pressure"]) 
    except KeyError:
        print("BMRB ID: ", key,"-->", "conditions: ", loc)
```

```python
pressure
```

```python
Counter(pressure).most_common()
```

```python
pressure_as_floats = []
for t in pressure:
    if t == "AMBIENT":
        pressure_as_floats.append(1)
    elif t == "NULL":
        pressure_as_floats.append(nan)
    elif t == "1 ATM":
        pressure_as_floats.append(1)
    elif t == "AMBIENT; AMBIENT":
        pressure_as_floats.append(1)
    elif t == "AMBIENT; AMBIENT; AMBIENT":
        pressure_as_floats.append(1)
    elif t == "1.0 ATM":
        pressure_as_floats.append(1)
    elif t == "AMBIENT; AMBIENT; AMBIENT;":
        pressure_as_floats.append(1)
    elif t == "1 ATM; 1 ATM":
        pressure_as_floats.append(1)
    elif t == "NULL; NULL":
        pressure_as_floats.append(nan)
    elif t == "1ATM":
        pressure_as_floats.append(1)
    elif t == "1 ATM; 1 ATM; 1 ATM; 1 ATM":
        pressure_as_floats.append(1)
    elif t == "ATMOSPHERIC ATM":
        pressure_as_floats.append(1)
    elif t == "1 ATM; 1 ATM; 1 ATM":
        pressure_as_floats.append(1)
    elif t == "1 ATMOSPHERE":
        pressure_as_floats.append(1)
    elif t == "1.0 ATM; 1.0 ATM":
        pressure_as_floats.append(1)
    elif t == "NULL; NULL; NULL; NULL; NULL;":
        pressure_as_floats.append(nan)
    elif t == "ATM":
        pressure_as_floats.append(1)
    elif t == "NULL; NULL; NULL":
        pressure_as_floats.append(nan)
    elif t == "1ATM; 1ATM":
        pressure_as_floats.append(1)
    elif t == "AMBIENT ATM":
        pressure_as_floats.append(1)
    elif t == "AMBIENT; NULL; NULL; NULL; NULL":
        pressure_as_floats.append(nan)
    elif t == "VARIABLE ATM":
        pressure_as_floats.append(nan)
    elif t == "1 BAR":
        pressure_as_floats.append(1)
    elif t == "1ATM ATM":
        pressure_as_floats.append(1)
    elif t == "1 BAR; 1 BAR":
        pressure_as_floats.append(1)
    elif t == "101300 PA":
        pressure_as_floats.append(1)
    elif t == "ATMOSPHERIC ATM; ATMOSPHERIC ATM":
        pressure_as_floats.append(1)
    elif t == "1 ATM; 1 ATM; 1 ATM; 1 ATM; 1":
        pressure_as_floats.append(1)
    elif t == "1.0 ; 1.0":
        pressure_as_floats.append(1)
    elif t == "1.0 ATM; 1.0 ATM; 1.0 ATM; 1.0":
        pressure_as_floats.append(1)
    elif t == "AMBINET":
        pressure_as_floats.append(1)
    elif t == "1.0 ATM; 1.0 ATM; 1.0 ATM":
        pressure_as_floats.append(1)
    elif t == "NULL; NULL; NULL; NULL":
        pressure_as_floats.append(nan)
    elif t == "NULL; NULL; NULL; NULL; NULL":
        pressure_as_floats.append(nan)
    elif t == "ATM; NULL":
        pressure_as_floats.append(nan)
    else:
        pressure_as_floats.append(float(t))
```

```python
pressure_as_floats_2 = []
for i in pressure_as_floats:
    if i > 200:
        print(i)
    else:
        pressure_as_floats_2.append(i)
```

```python
plt.figure(figsize=(15, 10))
plt.hist(pressure_as_floats, 20, density=False, alpha=0.75, histtype="stepfilled")
plt.show()
```
